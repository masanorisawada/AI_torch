```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def step_function(x):
    return np.where(x<=0, 0, 1)

x = np.linspace(-5,5)
y = step_function(x)

plt.plot(x,y)
plt.title('Step Function')
plt.xlabel('input')
plt.ylabel('output')
plt.show()
```

```python
def sigmoid_function(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-5,5)
y = sigmoid_function(x)

plt.plot(x,y)
plt.title('Sigmoid Function')
plt.xlabel('input')
plt.ylabel('output')
plt.show()
```

```python
def relu_function(x):
    return np.where(x<=0, 0, x)

x = np.linspace(-5,5)
y = relu_function(x)

plt.plot(x, y)
plt.title('ReLU Function')
plt.xlabel('input')
plt.ylabel('output')
plt.show()
```

```python
def leaky_relu_function(x):
    a = 0.01
    return np.where(x<=0, a * x, x)

x = np.linspace(-5,5)
y = leaky_relu_function(x)

z = x * 0
plt.plot(x, y, x, z, 'm--')
#plt.plot(x,y)
plt.title('Leaky ReLU Function')
plt.xlabel('input')
plt.ylabel('output')
plt.show()
```

```python
def identity_function(x):
    return x

x = np.linspace(-5,5)
y = identity_function(x)

plt.plot(x, y)
plt.title('Identity Function')
plt.xlabel('input')
plt.ylabel('output')
plt.show()
```

```python
def softmax_function(x):
    return np.exp(x)/np.sum(np.exp(x))

y = softmax_function(np.array([1,2,3]))

print('犬:{0:.2}、猫:{1:.2}、ハムスター{2:.2}'.format(y[0],y[1],y[2]))
print(np.sum(y))
```
